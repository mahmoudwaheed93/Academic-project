```python
import tkinter as tk
from tkinter import simpledialog, messagebox

def get_input(prompt):
    user_input = simpledialog.askstring("Input", prompt)
    if user_input is None or not user_input.lstrip('-').isdigit():
        messagebox.showerror("Error", "Invalid input. Please enter a valid positive number.")
        return None
    return int(user_input)

def show_output(output):
    messagebox.showinfo("Result", output)

def divisors():
    num = get_input("Enter a number:")
    if num is None:
        return
    result = ""
    for i in range(1, num + 1):
        if num % i == 0:
            result += f"{i}  {num // i}\n"
        else:
            result += f"{i} لا تقبل\n"
    show_output(f"Divisors of {num}: {result}")

def divisors2():
    num = get_input("Enter a number:")
    if num is None:
        return
    result = ""
    for i in range(1, int(num ** 0.5) + 1):
        if num % i == 0:
            result += f"{i} : {num // i}\n"
        else:
            result += f"{i} لا تقبل\n"
    show_output(f"Divisors of {num}: {result}")

def check():
    a = get_input("Enter the number a:")
    b = get_input("Enter the number b:")
    c = get_input("Enter the number c:")
    if a is None or b is None or c is None:
        return

    if b % a == 0 and c % a == 0:
        result = f"{a} divides {b} and {c}\n"
        if (b + c) % a == 0:
            result += f"{a} divides {b + c}"
        else:
            result += "كداب"
    else:
        result = "برجاء ادخال اعداد تقبل القسمه"

    show_output(result)

def check2():
    a = get_input("Enter the number a:")
    b = get_input("Enter the number b:")
    c = get_input("Enter the number c:")
    if a is None or b is None or c is None:
        return

    if b % a == 0 and c % b == 0:
        result = f"if {a} divides {b}\n"
        result += f"and {b} divides {c}\n"
        if c % a == 0:
            result += f"then {a} divides {c}"
        else:
            result += "كداب"
    else:
        result = "برجاء ادخال اعداد تقبل القسمه"

    show_output(result)

def remainder_positive():
    m = get_input("Enter the dividend (m) (positive):")
    n = get_input("Enter the divisor (n):")
    if m is None or n is None:
        return

    if m > 0 and n != 0:
        q = m // n
        r = m - q * n
        result = f"{m} = {q}({n}) + {r}\n"
        result += f"r = {r}"
    else:
        result = "Please ensure m is positive and n is not zero."

    show_output(result)

def remainder_negative1():
    m = get_input("Enter the dividend (m) (negative):")
    n = get_input("Enter the divisor (n):")
    if m is None or n is None:
        return

    if m < 0 and n != 0:
        q = abs(m) // n
        r = abs(m) - q * n
        r = n - r
        result = f"{m} = {q}({n}) + {r}\n"
        result += f"r = {r}"
    else:
        result = "Please ensure m is negative and n is not zero."

    show_output(result)

def remainder_negative2():
    m = get_input("Enter the dividend (m) (negative):")
    n = get_input("Enter the divisor (n):")
    if m is None or n is None:
        return

    if m < 0 and n != 0:
        q = -(m // n)
        r = m - (-q * n)
        result = f"{m} = {q}({n}) + {r}\n"
        result += f"r = {r}"
    else:
        result = "Please ensure m is negative and n is not zero."

    show_output(result)

def remainder_negative3():
    m = get_input("Enter the dividend (m) (negative):")
    n = get_input("Enter the divisor (n):")

    if m is None or n is None:
        return

    if m < 0 and n != 0:
        r = m
        while r < 0:
            r += n
        result = f"r = {r}"
    else:
        result = "Please ensure m is negative and n is not zero."

    show_output(result)

def check_prime():
    num = get_input("Check if the number is prime or not:")

    if num is None:
        return

    if num > 1:
        if num == 2:
            result = "prime"
        elif num % 2 == 0:
            result = "not prime"
        else:
            for i in range(3, int(num ** 0.5) + 1, 2):
                if num % i == 0:
                    result = "not prime"
                    break
            else:
                result = "prime"
    else:
        result = "not prime" if num == 1 else "negative number"

    show_output(result)

def gcd_common_divisors():
    a = get_input("Enter the first number (a):")
    b = get_input("Enter the second number (b):")
    if a is None or b is None:
        return
    divisors_a = [i for i in range(1, a + 1) if a % i == 0]
    divisors_b = [i for i in range(1, b + 1) if b % i == 0]
    common_divisors =list( set(divisors_a) & set(divisors_b))
    gcd = max(common_divisors) if common_divisors else 1
    show_output(f"GCD of {a} and {b}  = {gcd}")

def gcd_prime_factors():
    a = get_input("Enter the first number (a):")
    b = get_input("Enter the second number (b):")
    if a is None or b is None:
        return

    def prime_factors(num):
        factors = []
        divisor = 2
        while num > 1:
            while num % divisor == 0:
                factors.append(divisor)
                num //= divisor
            divisor += 1
        return factors

    factors_a = prime_factors(a)
    factors_b = prime_factors(b)

    common_factors = []
    unique_factors = set(factors_a) & set(factors_b)
    for factor in unique_factors:
        min_count = min(factors_a.count(factor), factors_b.count(factor))
        common_factors.extend([factor] * min_count)

    gcd = 1
    for factor in common_factors:
        gcd *= factor

    show_output(f"GCD of {a} and {b}  = {gcd}")

def gcd_successive_difference():
    a = get_input("Enter the first number (a):")
    b = get_input("Enter the second number (b):")
    if a is None or b is None:
        return
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
    show_output(f"GCD of the numbers = {a}")

def gcd_euclidean():
    m = get_input("Enter the first number (a):")
    n = get_input("Enter the second number (b):")
    if m is None or n is None:
        return
    while n != 0:
        r = m % n
        m = n
        n = r
    show_output(f"GCD of the numbers  = {m}")

def lcm_by_gcd():
    a = get_input("Enter the first number (a):")
    b = get_input("Enter the second number (b):")
    if a is None or b is None:
        return

    def gcd(a, b):
        while a != b:
            if a > b:
                a -= b
            else:
                b -= a
        return a

    gcd_value = gcd(a, b)
    lcm = (a * b) // gcd_value

    show_output(f"LCM of {a} and {b} = {lcm}")

def lcm_by_prime_factors():
    a = get_input("Enter the first number (a):")
    b = get_input("Enter the second number (b):")
    if a is None or b is None:
        return

    def prime_factors(n):
        factors = []
        divisor = 2
        while n > 1:
            while n % divisor == 0:
                factors.append(divisor)
                n //= divisor
            divisor += 1
        return factors

    factors_a = prime_factors(a)
    factors_b = prime_factors(b)

    unique_primes = list(set(factors_a + factors_b))
    lcm = 1

    for prime in unique_primes:
        power_a = factors_a.count(prime)
        power_b = factors_b.count(prime)
        max_power = max(power_a, power_b)
        lcm *= prime ** max_power

    show_output(f"LCM of {a} and {b}  = {lcm}")

root = tk.Tk()
root.title("Number Theory")
root.geometry("500x700")
root.config(bg="lightgray")

mylabel = tk.Label(root, text="Choose what do you want to calculate", font='arial 18 bold', fg="black", bg="lightgray")
mylabel.pack(pady=20)

# Buttons with a relaxed color palette
tk.Button(root, text="Divisors 1", command=divisors, bg="lightblue", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="Divisors 2", command=divisors2, bg="lightblue", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="(a b c) 1", command=check, bg="lightgreen", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="(a b c) 2", command=check2, bg="lightgreen", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="Remainder (positive)", command=remainder_positive, bg="lightyellow", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="Remainder (negative1)", command=remainder_negative1, bg="lightyellow", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="Remainder (negative2)", command=remainder_negative2, bg="lightyellow", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="Remainder (negative3)", command=remainder_negative3, bg="lightyellow", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="Prime or not", command=check_prime, bg="lightcoral", font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="GCD (Common Divisors)", command=gcd_common_divisors, bg='lightcoral', font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="GCD (prime factors)", command=gcd_prime_factors, bg='lightcoral', font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="GCD (Successive Difference)", command=gcd_successive_difference, bg='lightcoral', font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="GCD (Euclidean)", command=gcd_euclidean, bg='lightcoral', font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="LCM 1", command=lcm_by_gcd, bg='lightgreen', font='arial 14 bold', width=25, fg="black").pack(pady=5)
tk.Button(root, text="LCM (prime factors)", command=lcm_by_prime_factors, bg='lightgreen', font='arial 14 bold', width=25, fg="black").pack(pady=5)

root.mainloop()
```
